**Introduction**

 Karamoja is the most food-insecure region of Uganda. One of the main reasons is the low productivity level of the crops due to intense droughts as well as pest and disease outbreaks. Karamoja (Links to an external site.)Links to an external site.

In Karamoja, several NGOs provide technical support as well as farm inputs to the farmers experiencing extremely low yield. Though, they lack visibility into the overall state of the region and often need to rely on some very local sources of information to prioritize their activities.
Dalberg Data Insights (DDI) has been requested to develop a new food security monitoring tool to support the decision making of one of those NGOs active in Karamoja.

To do so, Dalberg Data Insights developed a methodology to remotely measure the yield of the two main staple crops of the region (i.e. sorghum and maize) based on satellite images. The agri-tech team just ran the model for the 2017 crop season.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your DataFrame (replace with your actual file path)
file_path = r'C:\Users\admin\Downloads\DATA\DATA\TABLES\Uganda_Karamoja_District_Crop_Yield_Population.csv'
df = pd.read_csv(file_path)

# Strip whitespace from column names (if necessary)
df.columns = df.columns.str.strip()

# Check if 'M_Prod_Tot' exists in the DataFrame
if 'M_Prod_Tot' in df.columns:
    # Calculate the mean of M_Prod_Tot
    mean_value = df['M_Prod_Tot'].mean()

    # Plotting the histogram for M_Prod_Tot
    plt.figure(figsize=(12, 6))
    plt.hist(df['M_Prod_Tot'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

    # Add a vertical line for the mean
    plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')

    # Add titles and labels
    plt.title('Histogram of Total Maize Production (M_Prod_Tot)', fontsize=16)
    plt.xlabel('Total Maize Production (M_Prod_Tot)', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add a legend
    plt.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()
```

**Corelation between Maize Yield and Sorghum Yield**

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Drop rows with missing values in either column
filtered_df = df[['M_Yield_Ha', 'S_Yield_Ha']].dropna()

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['M_Yield_Ha'], filtered_df['S_Yield_Ha'], alpha=0.5, label='Data Points')

# Fit a linear regression line
slope, intercept = np.polyfit(filtered_df['M_Yield_Ha'], filtered_df['S_Yield_Ha'], 1)
plt.plot(filtered_df['M_Yield_Ha'], slope * filtered_df['M_Yield_Ha'] + intercept, color='red', label='Regression Line')

# Adding titles and labels
plt.title('Scatter Plot of Maize Yield vs Sorghum Yield', fontsize=16)
plt.xlabel('Maize Yield (M_Yield_Ha)', fontsize=14)
plt.ylabel('Sorghum Yield (S_Yield_Ha)', fontsize=14)
plt.grid()
plt.legend()

# Annotate a few points (optional)
for i in range(0, len(filtered_df), max(1, len(filtered_df)//10)):  # Annotate every 10th point
    plt.annotate(f'({filtered_df["M_Yield_Ha"].iloc[i]:.2f}, {filtered_df["S_Yield_Ha"].iloc[i]:.2f})',
                 (filtered_df['M_Yield_Ha'].iloc[i], filtered_df['S_Yield_Ha'].iloc[i]),
                 textcoords="offset points",  # how to position the text
                 xytext=(0,10),  # distance from text to points (x,y)
                 ha='center', fontsize=8)

plt.show()
```

The maize yielded more than the Sorghum in Karamoja region
